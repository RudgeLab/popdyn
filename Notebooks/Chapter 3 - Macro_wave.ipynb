```python
from skimage.io import imread
import matplotlib.pyplot as plt
import numpy as np
from scipy.ndimage import gaussian_filter
```

```python
im = imread('/Users/timrudge/Pictures/10x_1.0x_MC4100_pLPT20&41_TiTweez_2_MMStack_Pos7.ome.tif')
im.shape
```

```python
frame = im[267,250:750,250:750,2]
plt.imshow(frame, origin='lower')
```

```python
gframe = gaussian_filter(frame, 16)
plt.imshow(gframe, origin='lower')
```

```python
gx,gy = np.gradient(gframe)
mgx = gx.mean()
mgy = gy.mean()
gmag = np.sqrt(mgx*mgx + mgy*mgy)
vx = -mgx / gmag
vy = -mgy / gmag
```

```python
plt.imshow(frame, origin='lower')
plt.plot([250, 250 + vx * 200], [250, 250 + vy * 200], 'w')
```
