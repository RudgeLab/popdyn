```python
import stackview
import imageio
import numpy as np
import tifffile as tiff
import os
```

```python
from skimage.io import imread, imsave
from skimage.transform import downscale_local_mean
import matplotlib.pyplot as plt
```

```python
path = '/Volumes/Expansion/Microscopy GY/2023/2023_02_13_German'
folder = os.listdir(path)
folder.sort()
folder
```

```python
for f in range(1,len(folder)):
    file_path = os.path.join(path,folder[f])
    print(f"File {file_path}")
    output_path = os.path.join(path,'downs_'+folder[f])
    im = imread(file_path)
    downsampled_stack = downscale_local_mean(im, (1, 2, 2, 1))
    with tiff.TiffWriter(output_path, bigtiff=True) as tif:
        for t in range(downsampled_stack.shape[0]):
            tif.write(
                downsampled_stack[t],
                photometric='rgb',
                compression='zlib',  # lossless compression
            )
    print(f"File t={f} saved")
```
